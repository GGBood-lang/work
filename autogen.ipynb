以下是我学习微软Autogen官网文档的知识总结

基础概念：
AutoGen 是一个开源编程框架，用于构建 AI 代理并促进多个代理之间的协作以解决任务，核心就是代理。 特点就是可以自动化实现写代码，画图等功能。通过大模型自动学习调整得到答案。

角色分配：
可以为代理分配角色，并让他们参与对话或互相聊天。 可以设置交流的轮数，max_turns=2 
以及可以设置结束终止词。  is_termination_msg=lambda msg: "good bye" in msg["content"].lower()

人机互动：
人工输入模式：NEVER，    TERMINATE(仅当满足终止条件时才要求人工输入)，       ALWAYS

对话模式：
双代理原理：
双代理聊天需要两个输入：一条消息，即呼叫者提供的字符串；一个上下文，指定聊天的各种参数。


聊天结束后，聊天历史记录将由聊天摘要器处理。summary_method="reflection_with_llm" 相当于用大模型对回答的问题进行文本摘要，默认情况下，它是聊天的最后一条消息。


连续聊天：
核心就是 将上一轮的文本摘要带入下一轮的上下文中



群聊：群聊的核心思想是所有代理都参与一个对话线程并共享相同的上下文。 群聊管理员是最重要的一环

每一个代理与群聊管理器开始双人聊天，群聊管理器在内部运行群聊，并在内部群聊结束时终止双人聊天。
